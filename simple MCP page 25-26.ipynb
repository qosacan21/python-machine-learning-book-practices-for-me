```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris

data = load_iris()

X = data.data
y = data.target
```

```python
print(X.shape)
print(y.shape)
```

```python
y = np.where(y == 1 ,1, -1)
```

```python
y
```

```python
class Perceptron(object):
    '''
    parameters:
    eta: float between 0 1 
    n_iter : 
    integer
    
    Attributes:
    w_ : 1d-array weights after fitting
    errors_: list. number of misclassication in each epoch
    '''
    
    def __init__(self, eta = 0.01, n_iter = 10):
        self.eta = eta
        self.n_iter = n_iter
        
    def fit(self,X,y):
        self.w_ = np.zeros(1 + X.shape[1]) # it will create array of 1+ n variable in 1d
        self.errors_ = []
        
        for _ in range(self.n_iter):
            errors = 0
            try:
                print(f'iteration_{_} :: {self.w_}')
            except:
                print('starting ...')
                
            for xi, target in zip(X,y):
                print(xi)
                print('---')
                update = self.eta * (target - self.predict(xi))
                self.w_[1:] += update * xi
                self.w_[0] += update
                errors += int(update != 0.0)
            self.errors_.append(errors)
        print('...finished')      
        return self
    
    def net_input(self, X):
        return np.dot(X, self.w_[1:]) + self.w_[0]
    
    def predict(self, X):
        return np.where(self.net_input(X) >= 0.0, 1, -1)
    
```

```python
model = Perceptron(eta = 0.1, n_iter = 10).fit(X,y)
```

```python
model.predict(X)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y,model.predict(X))
print('{:.3}%'.format(acc*100))
```
